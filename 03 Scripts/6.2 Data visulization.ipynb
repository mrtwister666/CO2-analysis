# correlation matrix

```python
# Create a correlation heatmap using matplotlib

corr = plt.matshow(df_clean.apply(lambda x: x.factorize()[0]).corr())
plt.show()
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_clean.apply(lambda x: x.factorize()[0]).corr()) # type of plot
plt.xticks(range(df_clean.shape[1]), df_clean.columns, fontsize=5, rotation=45) # x axis labels
plt.yticks(range(df_clean.shape[1]), df_clean.columns, fontsize=7) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=5) # add font size
plt.title('Correlation Matrix', fontsize=10) # add title
```

correlation was found between the following variables: 
- gun_stolen <> participant_age
- n_killed <> date
- address <> date
- gun_stolen <> n_guns_involved
- gun_type <> gun_stolen
